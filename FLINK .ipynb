### KPI's Requirements

```python
#Total sales
total_sales = df['Sales'].sum() 
# Average Sales
avg_sales = df['Sales'].mean()
# Number of Items Sold (no  Quantity Column,  so fallback = count of rows )
no_of_items_sold =df['Sales'].count()
#Average Rating (column name is 'Rating',not'Rating')
avg_ratings = df['Rating'].mean()
#Dislay
print(f"Total Sales: ${total_sales:,.0f}")
print(f"Average Sales: ${avg_sales:,.2f}")
print(f"Number of items sold: {no_of_items_sold:,.2f}")
print(f"Average rating:{avg_ratings:,.0f}")
```

```python
import matplotlib.pyplot as plt

sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

plt.pie(
    sales_by_fat,
    labels=sales_by_fat.index,
    autopct='%.1f%%',
    startangle =90
)
plt.title('Sales by Fat Content')
plt.axis('equal')
plt.show()
     
```

```python
fig, ax = plt.subplots()
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

ax.pie(
    sales_by_fat,
    labels=sales_by_fat.index,
    autopct='%.1f%%',
    startangle=90
)

# circle create for donut effect
centre_circle = plt.Circle((0,0), 0.5, color='white', fc='white')
fig.gca().add_artist(centre_circle)

plt.title('Method 2: Donut Chart With Center Circle')
plt.axis('equal')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group and sort sales by item type
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)


plt.figure(figsize=(10, 6))

# Create bar chart
bars = plt.bar(sales_by_type.index, sales_by_type.values, color='skyblue')

# Rotate x-axis labels for readability
plt.xticks(rotation=90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Type')

# Add value labels on top of each bar
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f'{bar.get_height():,.0f}',
        ha='center',
        va='bottom',
        fontsize=8
    )

# Adjust layout to prevent clipping
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

grouped =df.groupby(['Outlet Location Type','Item Fat Content'])['Sales'].sum().unstack()

grouped = grouped[['Low Fat','Regular']]

ax = grouped.plot(
    kind = 'bar',
    figsize=(8,5),
    title ='outlet tier by item fat content')
plt.xlabel('Outlet Location Tier')
plt.ylabel('Total Sales')
plt.legend(title='Item Fat Content')

plt.tight_layout()
plt.show()
    
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sales_grouped = (
    df
    .groupby(['Outlet Location Type', 'Outlet Size'])['Sales']
    .mean()
    .unstack()
)

plt.figure(figsize=(8,6))
sns.heatmap(sales_grouped, annot=True, fmt='.0f', cmap='coolwarm')
plt.title('Average Sales by Outlet Location Type and Outlet Size')
plt.xlabel('Outlet Size')
plt.ylabel('Outlet Location Type')
plt.show()
```

```python
# Adding a new column 
df['Region']=df['Outlet Location Type'].map({
    'Tier 1':'Urban',
    'Tier 2':'semi urban',
    'Tier 3': 'Rural'
})
```

```python
## Round Rating
df['Rating_round'] = df['Rating'].round(0)
sales_grouped = (
    df
    .groupby(['Region', 'Rating_round'])['Sales']
    .mean()
    .unstack()
)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(sales_grouped, annot=True, fmt='.0f', cmap='YlOrRd')
plt.title('Average Sales by Rating and Region')
plt.xlabel('Rating')
plt.ylabel('Region')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Line chart: Total Sales by Outlet Establishment Year
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

plt.figure(figsize=(9, 5))
plt.plot(sales_by_year.index, sales_by_year.values, marker='o', linestyle='-', color='teal')
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Outlet Establishment Year vs Total Sales')

# Add value labels
for x, y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x, y, f'{y:,.0f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()

```

```python
# Pie chart: Total Sales by Outlet Size
sales_by_size = df.groupby('Outlet Size')['Sales'].sum()

plt.figure(figsize=(4, 4))
plt.pie(
    sales_by_size,
    labels=sales_by_size.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['#FF9999', '#66B3FF', '#99FF99']  # Optional: custom colors
)
plt.title('Total Sales by Outlet Size')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Group and sort sales by outlet location type
sales_by_location = (
    df.groupby('Outlet Location Type')['Sales']
      .sum()
      .reset_index()
      .sort_values('Sales', ascending=True)
)

# Set figure size
plt.figure(figsize=(8, 3))

# Create horizontal bar chart
ax = sns.barplot(
    x='Sales',
    y='Outlet Location Type',
    data=sales_by_location,
    color='skyblue'
)

# Add title and labels
plt.title('Total Sales by Outlet Location Type')
plt.xlabel('Total Sales')
plt.ylabel('Outlet Location Type')

# Adjust layout
plt.tight_layout()
plt.show()
```
